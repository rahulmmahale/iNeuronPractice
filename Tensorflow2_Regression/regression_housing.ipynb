```python
import tensorflow as tf
import pandas as pd
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
```

```python
tf.__version__
```

```python
boston_housing_dataset = tf.keras.datasets.boston_housing
(train_x,train_y), (test_x, test_y) = boston_housing_dataset.load_data()
```

```python
print(train_x[0], train_y[0])
```

```python
scaler = StandardScaler()
scaler.fit(train_x)
```

```python
train_x_scaled = scaler.transform(train_x)
test_x_scaled = scaler.transform(test_x)
```

```python
print(train_x_scaled[0])
```

```python
bostonmodel1 = tf.keras.models.Sequential([
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(32,activation=tf.nn.relu,input_shape=[train_x.shape[1]]),
tf.keras.layers.Dense(64,activation=tf.nn.relu),
tf.keras.layers.Dense(1)
])
```

```python
optimiser = tf.keras.optimizers.Adam()
bostonmodel1.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

```python
trainedModel = bostonmodel1.fit(train_x_scaled, train_y, validation_split=0.2, epochs=100)
```

```python
bostonmodel1.evaluate(test_x_scaled, test_y)
```
